```python
import pandas as pd

data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'Date_of_Purchase': ['2023-12-12', '2023-12-15', '2023-12-18', '2023-12-20', '2023-12-24'],
    'Age': [25, 35, 45, 28, 31],
    'Annual_Income': [30000, 50000, 80000, 40000, 60000],
    'Purchase_Amount': [12000, 18000, 50000, 15000, 25000]
}

df = pd.DataFrame(data)
df['Date_of_Purchase'] = pd.to_datetime(df['Date_of_Purchase'])
df
```

```python
df['Income_Category'] = pd.cut(df['Annual_Income'],
                               bins=[0, 40000, 70000, 100000],
                               labels=['Low', 'Medium', 'High'])
df[['Annual_Income', 'Income_Category']]
```
